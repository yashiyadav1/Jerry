```python
#!pip install import_ipynb
import import_ipynb

import speech_recognition as sr
import pyttsx3
import requests
import json
from datetime import datetime
import random
import speech_recognition as sr
import pyttsx3
from datetime import datetime
import random
from news import get_news
from music import play_music
from email_utils import send_email
from weather import get_weather
from schedule import add_event
from reminder import set_reminder
from search import google_search

# Initialize the Text-to-Speech engine
engine = pyttsx3.init()

# Set the voice for the assistant
voices = engine.getProperty('voices')
engine.setProperty('voice', voices[0].id) # change the index to select a different voice

# Initialize the Speech Recognition engine
r = sr.Recognizer()

# Define a function to speak out the response
def speak(text):
    engine.say(text)
    engine.runAndWait()

# Define a function to listen to the user's input
def listen():
    with sr.Microphone() as source:
        print("Speak now...")
        r.pause_threshold = 1  # seconds of non-speaking audio before a phrase is considered complete
        audio = r.listen(source)

    try:
        print("Recognizing...")
        query = r.recognize_google(audio)
        print(f"User said: {query}\n")
    except Exception as e:
        print("Sorry, I did not understand that. Please try again.")
        query = None
    return query

# Define a function to get the current time
def get_time():
    now = datetime.now()
    current_time = now.strftime("%I:%M %p")
    return current_time

# Define the main function to run the assistant
def run_jerry():
    while True:
        query = listen()
        if query:
            if "hey" in query or "hello" or "hi" or "sup" in query:
                responses = ["Greetings, young Padawan.",
                             "Hello there."]
                response = random.choice(responses)
                speak(response)
            elif "time" in query:
                current_time = get_time()
                response = f"The current time is {current_time}"
                speak(response)
            elif "weather" in query:
                weather_info = get_weather()
                speak(weather_info)
            elif "news" in query:
                speak("Here are today's top headlines")
                get_news()
            elif "play" in query:
                play_music(query)
            elif "email" in query:
                speak("What is the subject of the email?")
                subject = listen()
                speak("What should be the content of the email?")
                body = listen()
                speak("Who is the recipient?")
                recipient = listen()
                send_email(subject, body, recipient)
            elif "schedule" in query:
                speak("What is the name of the event?")
                event_name = listen()
                speak("What is the start time of the event?")
                start_time = listen()
                speak("What is the end time of the event?")
                end_time = listen()
                add_event(event_name, start_time, end_time)
            elif "reminder" in query:
                speak("What task do you want to be reminded about?")
                task = listen()
                speak("At what time do you want to be reminded? Please specify in hours, minutes, and seconds.")
                time = listen()
                set_reminder(task, time)
            elif "bye" in query:
                responses = ["May the Force be with you.",
                             "Farewell, young Padawan.",
                             "Until next time."]
                response = random.choice(responses)
                speak(response)
                break
            else:
                search_result = google_search(query)
                response = f"Hmmm... Search for '{query}' I will. Found this, I have: {search_result}"
                speak(response)
                

if __name__ == "__main__":
    run_jerry()
```
